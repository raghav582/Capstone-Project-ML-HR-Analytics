Plot graph between satisfaction level and working hours.

```python
sns.scatterplot(dataset.average_montly_hours, dataset.satisfaction_level, hue=dataset.left)
```

There are three major category. Orange part showing employee who left.
Satisfaction level is high but working hour is also high.
Satisfaction level is low but working hour is high.
Satisfaction level is low but working hour is aslo low.

Effect of satisfaction level on exit.

```python
sns.barplot(dataset.left, dataset.satisfaction_level)
```

We can see people who left there average satisfactio level was low.

Effect of department on Exit.

```python
f, ax = plt.subplots(figsize=(10,5))
sns.countplot(dataset.Department, hue=dataset.left)
```

Major people who left are from sales, technical and support department.

```python
dataset.promotion_last_5years.unique()
```
